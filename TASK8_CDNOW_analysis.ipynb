```python
import pandas as pd
import numpy as np
```

```python
#读取文件，sep为分隔符，\s+为不确定个空格。header为是否把第一行当做列名。
#names给列命名，index_col 指出所索引列
data = pd.read_csv('CDNOW.txt',sep = '\s+',header = None,
                   names = ['user','date','quantity','price'],
                   index_col = ['user'])
```

```python
#查看是否有空值
data.isnull().sum()
```

```python
#各用户累计购买数量分析
data.groupby('user')['quantity'].sum()
```

```python
#各用户最大单笔购买数量
data.groupby('user')['quantity'].max()
```

```python
#各用户累计购买金额分析
data.groupby('user')['price'].sum()
```

```python
#各用户最大单笔消费金额
data.groupby('user')['price'].max()
```

```python
#各用户购买次数分析
data.groupby('user').count()
```

```python
#各用户第一次购买时间分析
data.groupby('user')['date'].min()
```

```python
#各用户最后一次次购买时间分析
period = data.groupby('user')['date'].max()
```

```python
#各用户最后一次消费与当前时间间隔
from dateutil.parser import parse
import datetime
per = []
for i in range(len(period)) :
    per.append (datetime.date.today() - datetime.datetime.strptime(str(period[i+1]), '%Y%m%d').date())
```
